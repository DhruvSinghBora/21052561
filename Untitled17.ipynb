```python
# Compute circle fit for the provided 3D points (assuming cylinder axis is parallel to Y,
# so we fit a circle in the X-Z plane). We'll show computed center, radius and residuals,
# and produce a quick matplotlib plot.
import numpy as np
import matplotlib.pyplot as plt

# Provided points (X, Y, Z) from the user's list
pts = np.array([
    [ 0.41261544,  0.21118298, -1.34184037],
    [ 0.37658763,  0.20642752, -1.32159299],
    [ 0.33659642,  0.2057943 , -1.31730623],
    [ 0.30171126,  0.21081642, -1.33476146],
    [ 0.4109301 ,  0.17171272, -1.34735126],
    [ 0.37479391,  0.16700348, -1.32662587],
    [ 0.33476181,  0.16662039, -1.32286072],
    [ 0.29935614,  0.17134249, -1.33904789],
    [ 0.40894944,  0.13277083, -1.35342554],
    [ 0.37298976,  0.1278865 , -1.33218569],
    [ 0.33316038,  0.12763918, -1.32783378],
    [ 0.29739446,  0.13221782, -1.34351158],
    [ 0.40706095,  0.09389536, -1.35951998],
    [ 0.37114355,  0.08914959, -1.33735613],
    [ 0.33138374,  0.08891006, -1.33320294],
    [ 0.29539319,  0.09336802, -1.34796282],
    [ 0.40525601,  0.05522897, -1.36511179],
    [ 0.36939957,  0.05046571, -1.34266423],
    [ 0.32961122,  0.05042805, -1.33794867],
    [ 0.2936153 ,  0.05482193, -1.35259233],
    [ 0.40350744,  0.01680495, -1.37088399],
    [ 0.36768722,  0.0124055 , -1.34781822],
    [ 0.3279425 ,  0.01227792, -1.3428213 ],
    [ 0.29199175,  0.01664512, -1.35690639],
    [ 0.40144424, -0.02149638, -1.37672016],
    [ 0.36606552, -0.02594798, -1.35316599],
    [ 0.32626368, -0.02571852, -1.3479375 ],
    [ 0.29049207, -0.02128669, -1.36178313]
])

# Extract X and Z (fit circle in X-Z plane)
X = pts[:,0]
Z = pts[:,2]

# Assemble linear system for circle: x^2 + z^2 + D*x + E*z + F = 0
A = np.column_stack([X, Z, np.ones_like(X)])
b = -(X**2 + Z**2)

# Solve least squares for [D, E, F]
sol, *_ = np.linalg.lstsq(A, b, rcond=None)
D, E, F = sol

# Center and radius
cx = -D/2.0
cz = -E/2.0
radius = np.sqrt(cx**2 + cz**2 - F)

# Residuals: distances from each point to the fitted circle
dists = np.sqrt((X - cx)**2 + (Z - cz)**2)
residuals = dists - radius
rms_residual = np.sqrt(np.mean(residuals**2))

# Print results
print(f"Fitted circle center in X-Z plane: (cx, cz) = ({cx:.6f}, {cz:.6f})")
print(f"Estimated radius: {radius:.6f}")
print(f"RMS residual (distance to circle): {rms_residual:.6f} (same units as input)")
print("\nResiduals for each point:")
for i, r in enumerate(residuals, start=1):
    print(f"P_{i}: {r:.6f}")

# Plot points and fitted circle
theta = np.linspace(0, 2*np.pi, 300)
circle_x = cx + radius * np.cos(theta)
circle_z = cz + radius * np.sin(theta)

plt.figure(figsize=(6,6))
plt.scatter(X, Z, label='Points (X,Z)', s=20)
plt.plot(circle_x, circle_z, label='Fitted circle', linewidth=1)
plt.scatter([cx], [cz], marker='+', s=80, label='Center (cx,cz)')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Z')
plt.title('Circle fit in X-Z plane (cylinder cross-section)')
plt.legend()
plt.grid(True)
plt.s
```

```python
import numpy as np

# Define the points
points = np.array([
    [0.41261544,  0.21118298, -1.34184037],
    [0.37658763,  0.20642752, -1.32159299],
    [0.33659642,  0.2057943 , -1.31730623],
    [0.30171126,  0.21081642, -1.33476146]
])

# Calculate pairwise distances between points
distances = np.linalg.norm(points[:, np.newaxis] - points, axis=2)

# Extract the unique distances (to avoid duplicates since distance matrix is symmetric)
unique_distances = distances[np.triu_indices_from(distances, k=1)]

# Average distance can be used as an estimate for the radius
radius_estimate = np.mean(unique_distances)
print(f"Estimated radius: {radius_estimate}")
```
